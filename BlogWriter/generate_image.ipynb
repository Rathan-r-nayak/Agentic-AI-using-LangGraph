```python
import urllib.parse
import requests
from langchain_core.tools import tool

@tool
def generate_image(prompt: str) -> str:
    """
    Call this tool when the user asks to create, draw, or generate an image.
    Pass in a highly detailed, descriptive visual prompt.
    """
    print(f"\n🎨 [Tool Execution] Generating free image for: '{prompt}'")
    
    # URL encode the prompt so it's safe to use in a web link
    encoded_prompt = urllib.parse.quote(prompt)
    
    # Pollinations creates images on the fly via URL
    image_url = f"https://image.pollinations.ai/prompt/{encoded_prompt}"
    
    try:
        # Download the image
        response = requests.get(image_url)
        if response.status_code == 200:
            filename = "free_agent_image.jpg"
            with open(filename, 'wb') as f:
                f.write(response.content)
            return f"Success! Image saved as '{filename}'."
        else:
            return "Failed to download the image."
            
    except Exception as e:
        return f"Failed to generate image. Error: {str(e)}"
```

```python
result = generate_image.invoke({
    "prompt": "A highly-detailed, simplified infographic diagram illustrating the data flow of a large language model (LLM). The process moves from left to right. On the far left: a user input text box showing a sentence being broken into 'tokens' (word chunks). These tokens flow into an 'embeddings' layer shown as dense vector coordinates. This flows into a multi-layer stack labeled 'TRANSFORMER ENCODER,' clearly visualizing multiple parallel 'SELF-ATTENTION HEADS' connecting tokens and calculating 'ATTENTION WEIGHTS' as thick glowing lines. A 'CONTEXT VECTOR' summaries this knowledge and feeds into a 'TRANSFORMER DECODER' stack. Finally, on the far right: 'VOCAB PROBABILITY DIST' leading to the sequential generation of 'GENERATED TOKENS' forming a complete response. Clean digital illustration style, precise lines, blue and orange color palette, professional diagrammatic layout."
})

print(result)
```

```python
import os
from huggingface_hub import InferenceClient
from langchain_core.tools import tool

@tool
def generate_technical_image(prompt: str) -> str:
    """Generates an image using the free Hugging Face API."""
    print(f"\n🎨 [Tool Execution] Generating image with FLUX for: '{prompt}'")
    
    # 1. Initialize the client (Replace with your actual token or use os.environ){}
    client = InferenceClient(api_key=os.environ.get('HUGGINGFACEHUB_API_TOKEN'))
    
    try:
        # 2. The SDK automatically handles the correct API routing and POST requests
        image = client.text_to_image(
            prompt=prompt,
            model="black-forest-labs/FLUX.1-dev"
        )
        
        # 3. Save the resulting image
        filename = "technical_diagram.png"
        image.save(filename)
        return f"Success! FLUX generated the image and saved it as '{filename}'."
        
    except Exception as e:
        return f"Failed to generate. Error: {str(e)}"
```

```python
generate_technical_image.invoke({"prompt":"A highly-detailed, simplified infographic diagram illustrating the data flow of a large language model (LLM)"})
```
