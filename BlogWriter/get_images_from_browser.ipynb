```python
import requests
from langchain_core.tools import tool

@tool
def fetch_local_images(query: str) -> list:
    """
    Searches the web for relevant image URLs.
    Returns a list of dictionaries with direct 'img_url' links.
    """
    print(f"🖼️ [Local Image Search] Querying: '{query}'")
    
    # 1. Point directly to your local Docker container's API endpoint
    url = "http://localhost:8080/search"
    
    # 2. Set up the exact parameters we want
    params = {
        "q": query,
        "categories": "images",
        "format": "json",
        "engines": "google images,bing images,duckduckgo images"
    }
    
    try:
        # 3. Make the request directly
        response = requests.get(url, params=params)
        response.raise_for_status() # Check for errors
        data = response.json()
        
        # 4. Extract exactly what we want without LangChain throwing it away
        extracted_images = []
        for result in data.get("results", []):
            if "img_src" in result:
                extracted_images.append({
                    "title": result.get("title", "No Title"),
                    "img_url": result.get("img_src"),      # The direct .jpg/.png link!
                    "source_url": result.get("url")        # The website it came from
                })
                
        # Return the top 3 images
        return extracted_images[:3]
        
    except Exception as e:
        return [f"Error fetching images: {str(e)}"]
```

```python
results = fetch_local_images.invoke({"query": "Udupi, karnataka"})

for res in results:
    print(f"Title: {res['title']}")
    print(f"Direct Image: {res['img_url']}")
    print("-" * 40)
```
