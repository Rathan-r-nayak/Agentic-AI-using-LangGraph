```python
from langgraph.store.memory import InMemoryStore
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from dotenv import load_dotenv
```

```python
load_dotenv()

embeddings = GoogleGenerativeAIEmbeddings(model="models/gemini-embedding-001")
```

```python
store = InMemoryStore(index= {"embed": embeddings, "dims": 768})
```

```python
namespace = ("user", "rathan")
```

```python
store.put(namespace, "1", {"data": "User likes python"})
store.put(namespace, "2", {"data": "User learning AI"})
store.put(namespace, "3", {"data": "User is learning machine learning"})
store.put(namespace, "4", {"data": "User dislikes overly theoretical explanations"})
store.put(namespace, "5", {"data": "User prefers step-by-step reasoning"})
store.put(namespace, "6", {"data": "User is based in India"})
store.put(namespace, "7", {"data": "User likes real-world analogies"})
store.put(namespace, "8", {"data": "User prefers bullet points over paragraphs"})
```

```python
namespace2 = ("user", "anand")

store.put(namespace2, "1", {"data": "User likes Java"})
store.put(namespace2, "2", {"data": "User learning spring"})
```

```python
store.get(namespace, "1")
```

```python
for item in store.search(namespace2):
    print(item)
```

```python
items = store.search(namespace, query="what user is learning", limit=1)

for item in items:
    print(item.value)
```

```python
items = store.search(namespace, query="what are user's preferences", limit=3)

for item in items:
    print(item.value)
```
